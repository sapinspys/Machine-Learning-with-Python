### Let's use Scikit-Learn

The first we need to do is split the data into an X array that holds the data to train our model on and a Y array with the target variable, which in this case is the 'price' column.

```python
df.columns
```

```python
# Features
X = df[['beds', 'baths', 'sq__ft']]

# Target Variable
y = df['price']
```

Next, we'll need to split our data into a training set for the model to learn on and a testing set to test our  model with later on. We also want to import the Linear Regression model from Scikit-Learn.

```python
from sklearn.model_selection import train_test_split
```

```python
'''
The following uses tuple unpacking to grab your training set and testing set

test_size is the percentage of our data set allocated to the test size, 40% here
random_state is optional, but ensures a specific set of random splits in our data'''

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

### Let's train our model!

```python
# Creating an instance of linear regression model
lnreg = LinearRegression()
```

```python
lnreg.fit(X_train, y_train)
```

```python
# Evaluate our model
print(lnreg.intercept_)
```

```python
# Grab the coefficients for each feature
lnreg.coef_
```

```python
X_train.columns
```

```python
# Organize the coefficients into a dataframe 
codf = pd.DataFrame(lnreg.coef_,X_train.columns, columns=['Coeff'])
```

```python
codf
```
